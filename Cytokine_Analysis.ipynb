## LDA

```python
from cytokine_analysis_fxns import mgh_LDA, mgh_LDA_varplot, mgh_LDA_loadings, mgh_LDA_biplot

mgh_LDA(mgh_data, label_col="COVID", n_components=1, vertical_jitter=True)

mgh_LDA_loadings(mgh_data, label_col="COVID", highlight=("IL6", "GDF15", "IL10", "IL1"))

mgh_LDA_loadings(mgh_data, label_col="COVID", top_n=30)

mgh_LDA_biplot(mgh_data, label_col="COVID")
```
